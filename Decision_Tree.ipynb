## Step 1. Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
```

## Step 2. Load the dataset

```python
data = data = {
    "Position": ["Junior", "Junior", "Senior", "Senior", "Manager", "Manager", "Senior Manager", "Director"],
    "Level": [1, 2, 3, 4, 5, 6, 7, 8],
    "Age": [21, 25, 30, 35, 40, 45, 50, 55],
    "Salary": [30000, 35000, 50000, 60000, 80000, 100000, 120000, 150000]
}
```

```python
df = pd.DataFrame(data = data)
```

```python
df.head()
```

## Step 3. Separate features and target variable

```python
X = df.iloc[:, 1:2].values # Selecting only the "Level" column
y = df.iloc[:, -1].values
```

Here, X contains the features used for prediction, and y contains the
target variable (salary).\
We select only the "Level" column as our feature
for simplicity.

## Step 4. Initialize and fit the Decision Tree Regressor

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X, y)
```

## Step 5. Visualize the Decision Tree Regression results

```python
X_grid = np.linspace(min(X), max(X), 100).reshape(-1, 1)
plt.scatter(X, y, color='red', label='Actual data')
plt.plot(X_grid, regressor.predict(X_grid), color='blue', label='Decision Tree Regression')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
